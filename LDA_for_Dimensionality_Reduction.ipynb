```python
# Import necessary libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = '/content/iris.csv'  # Replace with your file path
iris_df = pd.read_csv(file_path)

# Separating features and target labels
X = iris_df.drop('Species', axis=1)
y = iris_df['Species']

# Standardize the dataset
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply LDA to reduce to 2 components
lda = LDA(n_components=2)
X_lda = lda.fit_transform(X_scaled, y)

# Convert the result into a DataFrame for easier plotting
lda_df = pd.DataFrame(X_lda, columns=['LDA1', 'LDA2'])
lda_df['Species'] = y

# Plotting the LDA result
plt.figure(figsize=(8,6))
sns.scatterplot(x='LDA1', y='LDA2', hue='Species', data=lda_df, palette='bright')
plt.title('LDA: Dimensionality Reduction to 2 Components')
plt.xlabel('LDA1')
plt.ylabel('LDA2')
plt.grid(True)
plt.show()

# Apply PCA to reduce to 2 components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Convert the result into a DataFrame for easier plotting
pca_df = pd.DataFrame(X_pca, columns=['PCA1', 'PCA2'])
pca_df['Species'] = y

# Plotting the PCA result
plt.figure(figsize=(8,6))
sns.scatterplot(x='PCA1', y='PCA2', hue='Species', data=pca_df, palette='bright')
plt.title('PCA: Dimensionality Reduction to 2 Components')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.grid(True)
plt.show()
```
